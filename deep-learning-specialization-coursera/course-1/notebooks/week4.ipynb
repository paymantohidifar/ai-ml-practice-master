### Forward and backward propagration matrices dimensions

For training a $L$ layer neural network with $m$ examples and $n_x$ features, dimentions of the matrices are as following:

$$
\begin{equation}
\begin{aligned}
Z^{[l]}, A^{[l]}   &: (n^{[l]}, m) \\
dZ^{[l]}, dA^{[l]} &: (n^{[l]}, m) \\
W^{[l]}, dW^{[l]}  &: (n^{[l]}, n^{[l-1]}) \\
b^{[l]}, db^{[l]}  &: (n^{[l]}, m) \\
\end{aligned}
\label{eq:deep_nn_dims}
\end{equation}
$$

### Parameters vs Hyperparameteres

**Parameters:**
- $W^{[1]}$, $b^{[1]}$, $W^{[2]}$, $b^{[2]}$, $\text{...}$

**Hyperparameters:**
- Learning rate $\alpha$
- \# iterations
- \# Hidden layers ($L$)
- \# Size of a hidden layer ($n^{[1]}$, $n^{[2]}$, ...)
- Choice of the activation functions (e.g. ReLU, Leaky ReLU tanh, sigmoid, ...)

Hyperparameters control fate of parameters. That's why we call them "hyper" parameter.
